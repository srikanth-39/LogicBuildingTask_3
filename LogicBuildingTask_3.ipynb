```python
#Take the employees list
n=int(input("Enter the number of employees: "))
emp={}
for i in range(n):
  name=input("Enter the employee name:")
  score=int(input("Enter the employee score:"))
  emp[name]=score

#Check the highest score
highest=max(emp.values())
names=[name for name,score in emp.items() if highest==score]
print("Top Performers Eligible for Bonus:",",".join(names),f"(score:{highest})")
```

```python
#Take the user input
s=input("Enter the query:")

#Convert them into lower case and remove punctuation
s=s.lower()
import string
for ch in string.punctuation:
  s=s.replace(ch,"")

s=s.split()

d={}
for i in s:
  if i in d:
    d[i]+=1
  else:
    d[i]=1

c={ch:rep for ch,rep in d.items() if rep>1}
print(c)
```

```python
#Take the readings from the user
r=list(map(int,input("Enter the readings:").split()))
n=len(r)
res=[]
for i in range(n):
  if r[i]%2==0:
    res.append((i,r[i]))
print("Valid Sensor Readings (Hour, Value):")
print(res)
```

```python
#Take list of emails
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
g=0
y=0

for i in emails:
  if "gmail" in i:
    g+=1
  else:
    y+=1
n=len(emails)
gp=(g/n)*100
yp=(y/n)*100
print("gmail.com:",gp)
print("yahoo.com:",yp)
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
threshold = average * 1.3

spikes = [(day, value)
          for day, value in enumerate(sales, start=1)
          if value > threshold]

for day, value in spikes:
    print(f"Day {day}: {value}")

```

```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store counts
count_dict = {}

# Count frequency of each user ID
for user in user_ids:
    if user in count_dict:
        count_dict[user] += 1
    else:
        count_dict[user] = 1

# Display only duplicates
for user, count in count_dict.items():
    if count > 1:
        print(f"{user} → {count} times")
```
